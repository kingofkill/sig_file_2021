```python
import os
import dask
import math
import numpy as np
import pandas as pd
import seaborn as sns
import dask.dataframe as dd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from statsmodels.tsa.seasonal import seasonal_decompose
# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")

%matplotlib inline
```

```python
ecs_ali=pd.read_csv('./ECS_ali_cost.csv')
ecs_ali.rename(columns={'hw_c':'ecs_ali_hw','bw_c2':'ecs_ali_bw','all_cost':'ecs_ali_allcost'},inplace=True)
ecs_ali
```

```python
ecs_huw=pd.read_csv('./ECS_hw_cost.csv')
ecs_huw.rename(columns={'hw_c':'ecs_huw_hw','bw_c5':'ecs_huw_bw','all_cost':'ecs_huw_allcost'},inplace=True)
ecs_huw
```

```python
ens_ali=pd.read_csv('./ENS_aliu_cost.csv')
ens_ali.rename(columns={'hw_c':'ens_ali_hw','bw_c':'ens_ali_bw'},inplace=True)
ens_ali
```

```python
top_15=[64389,98051,78131,66394,48944,52024,80365,22399,31102,77787,64892,15301,83822,25295,13404,76645]
ens_ali_15=ens_ali[ens_ali['ali_uid'].isin(top_15)]
ecs_huw_15=ecs_huw[ecs_huw['ali_uid'].isin(top_15)]
ecs_ali_15=ecs_ali[ecs_ali['ali_uid'].isin(top_15)]
```

```python
ecs_ali_15
```

```python
ali_15 = pd.merge(ecs_ali_15,ens_ali_15,on='ali_uid')
all_15 = pd.merge(ali_15,ecs_huw_15,on='ali_uid')
all_15
```

```python
ecs_ali_hw=np.log((all_15['ecs_ali_hw']*0.15)).tolist()
ecs_ali_bw=np.log((all_15['ecs_ali_bw']*0.15)).tolist()

ecs_huw_hw=np.log((all_15['ecs_huw_hw']*0.15)).tolist()
ecs_huw_bw=np.log((all_15['ecs_huw_bw']*0.15)).tolist()

ens_ali_hw=np.log((all_15['ens_ali_hw']*0.15)).tolist()
ens_ali_bw=np.log((all_15['ens_ali_bw']*0.15)).tolist()
```

```python
ecs_ali_hw
```

```python
x = np.arange(len(top_15))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots(figsize = (12, 6))
ax.bar(x - width, ecs_huw_hw, width, label='HW Cloud_hw')
ax.bar(x, ecs_ali_hw, width, label='Ali Cloud_hw')
ax.bar(x + width, ens_ali_hw, width, label='Our Edge_hw')

ax.bar(x - width, ecs_huw_bw, width, bottom=ecs_huw_hw,label='HW Cloud_bw')
ax.bar(x, ecs_ali_bw, width, bottom=ecs_ali_hw,label='Ali Cloud_bw')
ax.bar(x + width, ens_ali_bw, width, bottom=ens_ali_hw,label='Our Edge_bw')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Monthly Cost($)')
#ax.set_title('Scores by group and gender')
ax.set_xticks(x)
#ax.set_xticklabels(labels)
ax.legend()
```

```python
all_15['save%ali']=(all_15['ecs_ali_allcost']-all_15['ens_cost'])/all_15['ecs_ali_allcost']*100
all_15['save%huw']=(all_15['ecs_huw_allcost']-all_15['ens_cost'])/all_15['ecs_huw_allcost']*100
```
