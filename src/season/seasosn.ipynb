```python
import os
import dask
import numpy as np
import pandas as pd
import seaborn as sns
import dask.dataframe as dd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
from statsmodels.tsa.seasonal import seasonal_decompose
# Ignore harmless warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
%matplotlib inline
```

```python
cols = ["ins_id", "region_id","ifnull(cpu_rate, 0)","report_ts"]
cpu = dd.read_csv('../../../data/June-all/T_INSTANCE_CPU/T_INSTANCE_CPU.csv',usecols=cols)
cpu = cpu.compute()
cpu.head()
```

```python
cpu.drop(['region_id'],axis=1,inplace=True)
#cpu.drop(['create_time'],axis=1,inplace=True)
cpu['date'] = pd.to_datetime(cpu['report_ts'],unit='s')
#因为时区不同，所以为了标准在原基础上对时间戳进行时间维度平移
cpu['date'] = pd.DatetimeIndex(cpu['date']) + timedelta(hours=8)
```

```python
cpu.head()
```

```python
def sea_stren(df):
    result = seasonal_decompose(df,freq=48, model='additive')
    #result.plot();
    result.resid.fillna(0, inplace=True)
    rs = result.seasonal + result.resid
    var_rs = np.var(rs)
    if(var_rs==0):
        y_sea = 16067
    else:
        y_sea = 1 - (np.var(result.resid)/var_rs)
    #y_sea = 1 - (np.var(result.resid)/var_rs)
    return y_sea
```

```python
vm_full = cpu.groupby([cpu['ins_id'],cpu['date']])[['ifnull(cpu_rate, 0)']].mean()

n=0

for i in vm_full.index.levels[0]:
    #bw_ins.loc[i].plot(figsize=(18,8),label = i);
    df = vm_full.loc[i]
    df_max = df.resample('30min').max()
    df_avg = df.resample('30min').mean()
    
    df_max.fillna(0, inplace=True)
    df_avg.fillna(0, inplace=True)
    
    if (len(df_max)< 864):
        continue
    new = pd.DataFrame({"vm":i, "sea_max":sea_stren(df_max['ifnull(cpu_rate, 0)'][:864]), "sea_avg":sea_stren(df_avg['ifnull(cpu_rate, 0)'][:864])},index=["0"])
    new.to_csv('./ali_30min_sea.csv',mode='a',header=0)
    n=n+1
    if(n%100==0):
        print('%d VMs have been processed'%(n))
```

```python
ali_30_sea=pd.read_csv('./ali_30min_sea.csv',names=['vm','sea_max','sea_avg'])
```

```python
azure_sea = pd.read_csv('./azure_30min_sea.csv',names=['vm','sea_max','sea_avg'])
azure_sea.head()
```

```python
ali_30_sea.fillna(0, inplace=True)
azure_sea.fillna(0, inplace=True)
ali_sea_max=ali_30_sea[ali_30_sea['sea_max']!=16067]
azu_sea_max=azure_sea[azure_sea['sea_max']!=16067]

sns.distplot(ali_sea_max['sea_max'],label='Our Edge',hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True})
sns.distplot(azu_sea_max['sea_max'],label='Azure Cloud',hist_kws={'cumulative': True, 'density': True}, kde_kws={'cumulative': True})
plt.title('Max cpu utilization seasonality')
plt.legend()
plt.xlabel('cpu seasonality')
```
